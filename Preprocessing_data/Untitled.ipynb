```python
import pandas as pd
import numpy as np

data = pd.read_csv(r'data/events2.csv')
```

```python
data.head()
data['timestamp'] = pd.to_datetime(data['timestamp'])
data_sorted = data.sort_values('timestamp')

# identifier for login logout (binary 1,0)
# find range for users and check if start or end points for other users is in this range
# find largest range and check if other users start/end times are within this range

login = data_sorted.loc[data_sorted['event_type']=='login',:]
logout = data_sorted.loc[data_sorted['event_type']=='logout',:]
len(logout)
len(login)
data_sorted.head()
```

```python
# login period will be start of data recording (for customers with logout first)
# logout period will be end of data recording (for customers with login last)
# add these values to each customer so we have 1-1 relationship. then merge login logout for each customer
# find first event_types and last_event types for each person
start_of_recording = min(data_sorted['timestamp'])
end_of_recording = max(data_sorted['timestamp'])

login_values = login['timestamp'].values
logout_values = logout['timestamp'].values

login_per_user = login.groupby('user_email').agg({'timestamp':'unique'}).reset_index()
logout_per_user = logout.groupby('user_email').agg({'timestamp':'unique'}).reset_index()
```
