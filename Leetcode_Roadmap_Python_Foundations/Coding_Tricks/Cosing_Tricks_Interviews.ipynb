# Coding Tricks
While progressing through the Leetcode roadmap, there are some quick tricks that I encountered for greater efficiency and may be asked in interviews.

## Duplicates

```python
def are_there_duplicates(nums):
    return len(nums) != len(set(nums))
x=[2,1,3,4,5,242,4,54,6,1,1]
are_there_duplicates(x)
```

# Check if Anagram
Anagram: a string that contains the exact same characters as another string, but the order of the characters can be different.

```python
s = "racecar"
t = "carrace"
def is_anagram_sort(str_1, str_2):
    return sorted(str_1) == sorted(str_2)
is_anagram_sort(s,t)
```
