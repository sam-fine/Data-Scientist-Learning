Design an algorithm to **encode** a list of strings into a single string,  
and then **decode** that string back to the original list of strings.

### Example 1

**Input:**
```python
["neet", "code", "love", "you"]
```
**Eg mid Output:**
```python
["neet:code:love:you"]
```
**Final output is same as input**
```

O(n)

````python
strs = ["neet", "code", "hi", "you", "!@#$%^&*()"]
a = "!@#$%^&*()"
len(a)
````

````python
def encode(strs):
    '''
    Make delimiter the length of the word and any character after this. 
    When finding the length of the word for the decode, we will stop when we have the character.
    '''
    res = ''
    for s in strs:
        res += str(len(s)) + ':' + s
    return res
    
def decode(en_str):
    original_arr = []
    current = 0
    i = 0
    while i< len(en_str):
        j = i
        while en_str[i] != ':':
            i+=1
        word_size = int(en_str[j:i])
        new_start = i+1+word_size
        original_arr.append(en_str[i+1:new_start])
        i = new_start
    return original_arr
        
   #return en_str.split(' ')
decode(encode(strs))
````
