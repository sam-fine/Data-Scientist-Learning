# Anagram (String manipulation)
Anagram: a string that contains the exact same characters as another string, but the order of the characters can be different.

```python
s = "racecar"
t = "carrace"
```

## 1. Using Dictionary (HashMap)

**Complexity:**
- **Time:** O(s+t) as iterate through 2 dictionaries of max size s and t
- **Space:** O(s+t) two dictionaries of size s and t

```python
def is_anagram_dict(str_1, str_2):
    if len(str_1) != len(str_2):
        return False
    s_dict, t_dict = {}, {}
    for i in range(len(s)):
        s_dict[s[i]] = 1 + s_dict.get(s[i],0) # returns 0 if the key does not exist
        t_dict[t[i]] = 1 + t_dict.get(t[i],0)
    return s_dict == t_dict
is_anagram_dict(s,t)
```

Same as above but using the Counter function

## 2. Sort

**Complexity (depends on the sorting algorithm):**
- **Time:** Depends on sorting algorithm (n or nlogn)
- **Space:** O(1) no array stored

```python
def is_anagram_sort(str_1, str_2):
    return sorted(str_1) == sorted(str_2)
is_anagram_sort(s,t)
```
