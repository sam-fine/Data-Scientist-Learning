Given an array of strings `strs`, group all **anagrams** together into sublists.  
You may return the output in any order.

An **anagram** is a string that contains the exact same characters as another string,  
but the order of the characters can be different.

## **1. Sorting and Appending to Dictionary**  
**Complexity:**
- **Time:** O(m*nlogn) iterate through array and sorting algorithm nlogn each time
- **Space:** O(m*n) as store each string once and n total character storage.

Where m is the number of strings and n is the length of the longest string.

````python
from collections import defaultdict
def groupAnagrams(strs):
    '''
    Each anagram is equal when the word is sorted.
    Append to dict keys the words that are the same when sorted.
    '''
    res = defaultdict(list) # so list is created in val if no key present
    for word in strs:
        key = ''.join(sorted(word)) # sort the word and make it a string
        res[key].append(word)
    return list(res.values())
groupAnagrams(strs)
````
