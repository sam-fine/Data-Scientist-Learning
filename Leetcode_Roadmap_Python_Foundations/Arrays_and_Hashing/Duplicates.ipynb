**Goal:** Detect, count, or remove repeated elements in a list using Python fundamentals.

## **1. Using Set (simple check)**  
**Complexity:**
- **Time:** O(n) as building a set iterates through all elements once.
- **Space:** O(n) as storing elements in a set which has max n elements.

```python
x = [1,2,3,2,1,5,6,7]
def has_duplicates(nums):
    return len(nums) != len(set(nums))
has_duplicates(x)
```

## 2. Using Set (early exit)

**Complexity:**
- **Time:** O(n) worst case, O(k) best case (k=first duplicate).
- **Space:** O(n) storing seen elements.

```python
def has_duplicates_set_early(nums):
    seen = set()
    for n in nums:
        if n in seen:
            return True
        seen.add(n)
    return False
has_duplicates_set_early(x)
```

## 3. Using Counter (Get all duplicates / frequency)
Return the duplicates.
**Complexity:**
- **Time:** O(n) iterates through list to count the frequency
- **Space:** O(n) storing counts in dict

```python
from collections import Counter

def has_duplicates_counter(nums):
    counts = Counter(nums)
    print(counts)
    duplicates = [num for num, count in counts.items() if count>1]
    return duplicates
has_duplicates_counter(x)
```
