Given an array of integers `arr` and an integer `target`, return the indices `i` and `j` such that:


$
arr[i] + arr[j] = target \quad 
\text{and} \quad i \neq j
$

You may assume that every input has **exactly one** pair of indices `i` and `j` that satisfy the condition.  
Return the answer with the **smaller index first**.

**Input:**

arr = [3, 4, 5, 6]

## **1. Brute Force**  
**Complexity:**
- **Time:** O($n^2$) as we have $(n-1) + (n-2) + ... = \frac{n(n-1)}{2}$ iterations.
- **Space:** O(1) as return [i,j]

````python
def two_sum_brute_force(arr, target: int):
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if i==j:
                continue
            elif arr[i] + arr[j] == target:
                return [i,j]
    return False
two_sum(arr, target)
````

## **2. Dictionary (HashMap)**
Iterate through each element, checking if target-element is in the dictionary of previous elements. If it is, return indexes of element and diff. If not, then we add the element to the dictionary with it's index.   
**Complexity:**
- **Time:** O(n) as iterating through array
- **Space:** O(n) as may have to add all elements in array to dictionary.

````python
def two_sum_other(arr, target: int):
    '''
    Iterate through each element, checking if target-element is in the dictionary of previous elements.
    If it is, return indexes of element and diff. 
    If not, then we add the element to the dictionary with it's index. 
    '''
    prevMap = {} # val : index
    for i, n in enumerate(arr):
        diff = target - n
        if diff in prevMap:
            return [prevMap[diff], i]
        prevMap[n] = i
    return False
two_sum_other(arr, target)

````
