# ðŸ”¢ Top K Frequent Elements

Given an integer array `nums` and an integer `k`, return the **k most frequent elements** within the array.

The test cases are generated such that the answer is always unique.  
You may return the output in **any order**.

---

### Example 1

**Input:**
```python
nums = [1, 2, 2, 3, 3, 3]
k = 2
```
```python
result = [2,3]

````python
nums = [1, 2, 2, 3, 3, 3,3,5,5,5,5]
k = 2
````

````python
def fun(nums,k):
    counts = {}
    res = []
    for val in nums:
        counts[val] = 1 + counts.get(val,0)
    count_arr = [[] for i in range(len(nums)+1)]
    for key,count in counts.items():
        count_arr[count].append(key)
    for l in range(len(nums),-1,-1):
        res.extend(count_arr[l])
        if len(res)>=k:
            return res
    return res
fun(nums,k)       
````

````python
from collections import Counter
def k_most(nums, k):
    counts = Counter(nums)
    # counts = {}
    # for num in nums:
    #     counts[num] = 1 + counts.get(num,0)

    # counts = defaultdict(int)
    # for val in nums:
    #     counts[val] += 1
    
    freq = [[] for i in range(len(nums)+1)] # count of each number in list is never 0 as exists in list
    for key, count in counts.items():
        freq[count].append(key) # list with counts as index and list of elements with corresponding counts as values

    res = []
    for j in range(len(freq)-1,0,-1): # start from end of list as this is the highest count
        res.extend(freq[j]) # all elements in list added to res
        if len(res) ==k:
            return res
    return res
        
k_most(nums,k)    
````

````python
from collections import Counter
def k_most(nums, k):
    counts = Counter(nums)
    if len(counts)<=k:
        return list(counts.keys())
    top_vals = sorted(list(counts.values()), reverse=True)[:k]
    fin = []
    for i in set(nums):
        if counts[i] in top_vals:
            fin.append(i)
    return fin

nums = [2]
k_most(nums,k)    
````
