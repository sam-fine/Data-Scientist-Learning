# Quick Sort

- recursive
- pivot
  - Is in the correct position in final, sorted array
  - Items to the left are smaller
  - Items to the right are larger

- Move pivot to end of array
- Find itemFromLeft: first item from the left of array that is larger than pivot
- Find itemFromRight: first item from the right of array that is smaller than pivot
- swap itemfrom left with itemfrom right
- keep swapping until itemfromleft has a greater index than item from right
- swap item from left with pivot and pivot is in correct index
- Now repeat but with each subset to the left and right of the pivot

How to choose the pivot?
- want to split the array in half to even out the workload
- eg median of three. look at first, middle and last elements of the array, sort them and choose the middle item as the pivot

```python
def quicksort(arr):
    if len(arr)>1:
        pivot = arr.pop()
        grt, equal_arr, sml = [], [pivot], []
        for item in arr:
            if item == pivot:
                equal_arr.append(item)
            elif item < pivot:
                sml.append(item)
            else:
                grt.append(item)
        return quicksort(sml) + equal_arr + quicksort(grt)
    else:
        return arr
```

```python
x=[5,6,2,4,5,7,-2,3,-65,0]
quicksort(x)
```
