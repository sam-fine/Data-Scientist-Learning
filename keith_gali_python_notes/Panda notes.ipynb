Make changes to data

```python
##add column:eg lets total all the stats for each pokemon(HP+attack+defence etc)
df['Total']= df['HP'] + df['Attack']+ df['Defense']+ df['Sp. Atk']+df['Sp. Def']+df['Speed']
df.head()

##drop columns
df = df.drop(columns=['Total'])
df.head()

##add columns different way
df['Total']= df.iloc[:, 4:10].sum(axis = 1) # :, means all rows 4:10 (not including 10) means 4th to 10th column
df.head() # sum(axis = 1) means summing horizontally, axis = 0 means vertically

##change order of columns
#make columns into one list
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
print(cols[0:4]) 
print(cols[-1]) # note is a string not a list in this case as is one word
```

Saving our Data (into our directory)

```python
##save our modified data set in a csv
df.to_csv('modified.csv', index = False)# got rid of index column
```

Filtering our Data

```python
new_df = df.loc[(df['Type 1']== 'Grass') & (df['Type 2'] == 'Poison') & (df['HP']>70)] # grass and poison columns

##Note, in panda data frame, use & not 'and', | not 'or'

#save new data frame

new_df.to_csv('filtered.csv')

#reset index 
#new_df = new_df.reset_index(drop = True) # drops old index column
# can use inplace to automaticlly assign the new df to 'new_df'
new_df.reset_index(drop = True, inplace = True)

##contains
##filter out data that contains something
#eg filter out data that has 'Mega' in the name
not_mega = df.loc[~df['Name'].str.contains('Mega')] # use string as Name is strings

#Note: ~ means 'not' in panda
```

Conditional Changes

```python
##Change the data 'Fire' to 'Flamer'
df.loc[df['Type 1']=='Fire', 'Type 1'] = 'Flamer'

#make all Flamer pokemon legendary

df.loc[df['Type 1']=='Flamer', 'Legendary'] = True # legendary column changes to True

# df has been changed so make it back to original (modified)

df = pd.read_csv('modified.csv')
```

Aggregate Statistics (groupby)

```python
## mean of Type 1 statistics and sort

df.groupby(['Type 1']).mean().sort_values('Defense', ascending = False) #Steel has highest defense

#sum Type 1 Statistics
df.groupby(['Type 1']).sum()

#count how many of each Type 1 there are

df.groupby(['Type 1']).count()['#']

# how many of each Type 2 are in each type 1

df.groupby(['Type 1']).count()

```
