```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Basic Graph

```python
#plot (1,2), (2,4), (3,6)

#resize graph
plt.figure(figsize = (10,5))



#define x and y arrays
x=[1,2,3,4]
y=[2,4,6,8]

#first line
plt.plot(x,y, label = 'y=2x', color = 'orange', linewidth = 3, marker ='.' ,markersize = 10)  #need label for the legend 

#second line
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'r--', label = 'y=x^2') #used shorthand
plt.plot(x2[5:], x2[5:]**2, 'r') # change line style in intervals

#Note: for colors, use any hex value, eg '#abcdef'. look up colour picker on google


#label graph
plt.title('First Graph!', fontdict ={'fontname':'Comic sans MS', 'fontsize': 20}) #fontdict- change font/size etc
plt.xlabel('x axis', fontdict = {'fontsize':20})
plt.ylabel('y label',fontdict = {'fontsize':20})

plt.legend()

plt.xticks([0,1,2,3,4], rotation = 45)
plt.yticks([0,2,4,6,8,10])

plt.show()


#note: can use shorthand
# format =fmt ='[color],[marker],[line]'  plt.plot(x,y,fmt)
# eg: plt.plot(x,y,'r.--' ,markersize = 10 ,label = 'y=2x')
# cpl - colour,point,label   so like pcr test but l at end
```

## Bar chart

```python
labels = ['A','B','C']
values =[1,4,2]

bars = plt.bar(labels,values)

bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
plt.show()
```
